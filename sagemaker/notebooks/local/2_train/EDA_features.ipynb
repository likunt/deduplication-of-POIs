## 0. Import libraries

```python
# scatter plot
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import pandas as pd
df = pd.read_csv("../../1_datasets_train_fs/train_all_region123/pairs_fs_features.csv", low_memory=False)

x = []
y = []
colors = []
for _, row in df.iterrows():
    name_jaro = row['name_jaro']
    name_embed = row['name_embed']
    addr_jaro = row['addr_jaro']
    addr_embed = row['addr_embed']
    dist_eucd = row['dist_eucd']
    if dist_eucd>0.0005:
        continue
    x.append(name_jaro)
    y.append(addr_jaro)
    if row['match']==1:
        c = 'g'
    else:
        c = 'm'
    colors.append(c)

plt.scatter(x, y, c=colors, alpha = 0.5)
plt.show()
```
